```python
import pandas as pd

hemorrhage_dataset_path = "dataset/hemorrhage_diagnosis.csv"
hemo_data = pd.read_csv(hemorrhage_dataset_path)
hemo_data
```

```python
hemo_data["has_hemorrhage"] = (
    (hemo_data["Intraventricular"])
    | (hemo_data["Intraparenchymal"])
    | (hemo_data["Subarachnoid"])
    | (hemo_data["Epidural"])
    | (hemo_data["Subdural"])
)
cols = ["Intraventricular", "Intraparenchymal", "Subarachnoid", "Epidural", "Subdural"]
total = 0
print(f"total slices with hemorrhage: {(hemo_data["has_hemorrhage"] == 1).sum()}")
```

```python
hemo_data
```

```python
# a patient has an hemorrhage if at least one of their slices has any type of hemorrhage
has_hemorrhage_per_patient = (
    hemo_data["has_hemorrhage"].groupby(hemo_data["PatientNumber"]).sum() > 0
).astype(int)
has_hemorrhage_per_patient.groupby(has_hemorrhage_per_patient).count()
```
